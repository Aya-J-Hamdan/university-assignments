```python
import pandas as pd
import numpy as np
from google.colab import files

uploaded=files.upload()
for file_name in uploaded.keys():
  df=pd.read_csv("heart.csv")

duplicate_count=df.duplicated().sum()
missing_count=df.isnull().sum().sum()
if missing_count>0:
  numeric_columns=df.select_dtypes(include=np.number).columns
  df_filled=df.copy()
  for col in numeric_columns:
    df_filled[col]=df[col].fillna(df[col].mean())
else:
    df_filled=df.copy()

for col in df_filled.columns:
  if df_filled[col].dtype==object:
    df_filled[col]=pd.to_numeric(df_filled[col], errors='coerce')

df_values=df_filled.values
min_vals=np.nanmin(df_values, axis=0)
max_vals=np.nanmax(df_values, axis=0)

df_normalized=(df_values - min_vals)/(max_vals - min_vals)

df_normalized=pd.DataFrame(df_normalized, columns=df.columns)

print("Duplicate data count:", duplicate_count)
print("missing data count:", missing_count)
print("normalized data with missing values filled with mean:", df_normalized)
```
